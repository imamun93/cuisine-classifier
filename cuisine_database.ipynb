```python

epicurious_cuisines = ['african', 'american', 'british', 'french', 'greek', 
                       'indian', 'irish', 'italian', 'japanese', 'korean', 'mexican',
                      'thai', 'turkish']
ingredients_list = []
name_ingredients_cuisine = []

for cuisine in epicurious_cuisines:
    cuisine_get = requests.get(f'https://www.epicurious.com/cuisine/{cuisine}')
    cuisine_text = cuisine_get.text
    cuisine_soup = BeautifulSoup(cuisine_text)
    cuisine_recipes_list = cuisine_soup.findAll('div', {'class':"list_c_hed"})
    recipes = [recipe.findNext('a').get('title') for recipe in cuisine_recipes_list]
    links = [recipe.findNext('a').get('href') for recipe in cuisine_recipes_list]
    if cuisine == 'thai':
        cuisine = "thai_and_south-east_asian"
    if cuisine == 'turkish':
        cuisine = "turkish_and_middle_eastern"

    for link in links:
        recipe = requests.get(f'https://www.epicurious.com{link}')
        recipe_text = recipe.text
        recipe_soup = BeautifulSoup(recipe_text)
        if recipe_soup.findAll('li', {'class':"ingredient"}):
            recipe_ingredients_list = recipe_soup.findAll('li', {'class':"ingredient"})
            ingredients = ' '.join([ingredient.text for ingredient in recipe_ingredients_list])
            if len(ingredients) > 1:
                ingredients = ingredients
        if recipe_soup.find('h1').text:
            name = recipe_soup.find('h1').text
        n_i_c = (name, ingredients, cuisine)
        name_ingredients_cuisine.append(n_i_c)
```
