```python
import pandas as pd
first_database = pd.read_csv('all_recipe_data.csv')
```

```python
first_database = first_database.rename(index=str, columns={"Recipe_Name": "recipe_name", "Recipe_Ingredients": "recipe_ingredients", "Cuisine":"cuisine"})
first_database = first_database.drop_duplicates(subset='recipe_name')
```

```python
second_database = pd.read_csv('cuisines.csv')
second_database = second_database.rename(index=str, columns={'0': "recipe_name", '1': "recipe_ingredients", '2':"cuisine"})

second_database.head()
```

```python
chinese_database = pd.read_csv('chinese_recipes.csv')
chinese_database = chinese_database.rename(index=str, columns={'0': "recipe_name", '1': "recipe_ingredients", '2':"cuisine"})

chinese_database.head()
```

```python

full_database = pd.concat([first_database, second_database, chinese_database])
full_database = full_database.drop_duplicates(subset='recipe_ingredients')

# df.describe()
```
